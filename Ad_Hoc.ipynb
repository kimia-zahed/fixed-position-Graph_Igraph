```python
import numpy as np
import matplotlib.pyplot as plt
import igraph as ig
```

```python
l= np.int16(65) #length of feild
n= np.int16(100) #number of node
mean_radius= np.float16(10)
```

```python
def Node_make(x,y,r):
    Nodes=np.array([],dtype=np.float16)
    for i in range(n):
        Nodes=np.append(Nodes,((x[i],y[i]),r[i]))
    Nodes=Nodes.reshape((n,2))
    return Nodes
```

```python
def make_Edges(x,y,r):
    #cs=np.array([])
    Edges=np.array([],dtype=np.float16)
    counter=0
    for i in range(n):
        counter=i+1
        for j in range(counter,n,1):
            c= np.sqrt((x[i]-x[j])**2 +(y[i]-y[j])**2,dtype= np.float16)
            #cs=np.append(cs,c)
            shart=r[i]
            if r[j]<=r[i]:
                shart=r[j]
            if (c<=shart):
                Edges=np.append(Edges,[i,j])
    #a=np.int16(len(Edges)/2)
    #Edges= Edges.reshape((a,2))      
    return Edges
```

```python
mu, sigma = 0, 1 # mean and standard deviation
R= mean_radius +np.float16(np.random.normal(mu, sigma, n))
```

```python
x_cord= np.float16(np.random.uniform(0,l,n))
y_cord= np.float16(np.random.uniform(0,l,n))
```

```python
nodes=Node_make(x_cord,y_cord,R);
#nodes
```

```python
#plt.plot(x_cord,y_cord,'o')
```

```python
edges=make_Edges(x_cord,y_cord,R)
```

```python
g=ig.Graph()
g.add_vertices(n)
```

```python
le=int(len(edges))
for i in range(0,le,2):
    e1=int(edges[i])
    e2=int(edges[i+1])
    g.add_edges([(e1,e2)])
    #print(edges[i])
```

```python
ig.plot(g)
```
